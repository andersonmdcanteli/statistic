```python
fig, (axs1, axs2) = plt.subplots(1,2)
fig.set_size_inches(16,8)
fig.suptitle("Comparando as edições")
axs1.scatter(cachorro_peso, cachorro_altura, label="Cachorros")
axs1.scatter(gatos_peso, gatos_altura, label="Gatos")
axs1.scatter(pinguins_peso, pinguins_altura, label="Pinguins")
axs1.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Base")
axs1.legend()

axs2.scatter(cachorro_peso, cachorro_altura, label="Cachorros")
axs2.scatter(gatos_peso, gatos_altura, label="Gatos")
axs2.scatter(pinguins_peso, pinguins_altura, label="Pinguins")
axs2.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Alterando o tamanho da fonte da legenda")
axs2.legend(fontsize='xx-large')
plt.show()
```

<br id="cor-fonte">

<p style="color: mediumblue; text-align: justify; font-size: 22px; font-family: Times New Roman; font-weight: bold;">
    Cor da fonte
</p>

<br>

<p style="text-align: justify; font-size: 18px; font-family: Times New Roman;">
&emsp;&emsp;A cor da fonte pode ser alterada passando uma <i><strong>string</strong></i> com o nome da cor, ou uma <i><strong>list</strong></i> contendo <i><strong>strings</strong></i> com o nome da cor, para o parâmetro <i><strong>labelcolor</strong></i>:
</p>

<br>

<p style="text-align: center; color:red; font-weight: bold; font-family: Times New Roman;">ATENÇÃO</p>

<p style="text-align: center; font-size: 18px; font-family: Times New Roman;">
O <i><strong>labelcolor</strong></i>  funciona apenas na versão 3.2 ou superior do matplotlib
</p>


<br>
<p style="text-align: right; font-size: 12px; font-family: Times New Roman;"><a href="#topicos"><i>voltar ao índice</i></a></p>

```python
fig, (axs1, axs2) = plt.subplots(1,2)
fig.set_size_inches(16,8)
fig.suptitle("Comparando as edições")
axs1.scatter(cachorro_peso, cachorro_altura, label="Cachorros")
axs1.scatter(gatos_peso, gatos_altura, label="Gatos")
axs1.scatter(pinguins_peso, pinguins_altura, label="Pinguins")
axs1.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Base")
axs1.legend()

axs2.scatter(cachorro_peso, cachorro_altura, label="Cachorros")
axs2.scatter(gatos_peso, gatos_altura, label="Gatos")
axs2.scatter(pinguins_peso, pinguins_altura, label="Pinguins")
axs2.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Alterando a cor do texto da legenda com uma única cor")
axs2.legend(labelcolor='red')
plt.show()
```

```python
fig, (axs1, axs2) = plt.subplots(1,2)
fig.set_size_inches(16,8)
fig.suptitle("Comparando as edições")
axs1.scatter(cachorro_peso, cachorro_altura, label="Cachorros")
axs1.scatter(gatos_peso, gatos_altura, label="Gatos")
axs1.scatter(pinguins_peso, pinguins_altura, label="Pinguins")
axs1.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Base")
axs1.legend()

axs2.scatter(cachorro_peso, cachorro_altura, label="Cachorros")
axs2.scatter(gatos_peso, gatos_altura, label="Gatos")
axs2.scatter(pinguins_peso, pinguins_altura, label="Pinguins")
axs2.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Utilizando uma lista com cores pré-definidas para alterar a cor da legendas")

cores = ['red', 'green', 'blue']

axs2.legend(labelcolor=cores)
plt.show()
```

<br>

<p style="text-align: justify; font-size: 18px; font-family: Times New Roman;">
    Também podemos passar três <i><strong>strings</strong></i> de referência para coincidir com as cores do marcador:
</p>
<ul style="font-size: 18px; font-family: Times New Roman;">
    <li>'linecolor': o texto terá a mesma cor que a cor da;</li>
    <li>'markerfacecolor' ou 'mfc': o texto terá a mesma cor que a cor de dentro do marcador;</li>
    <li>'markeredgecolor' ou 'mec': o texto terá a mesma cor que a cor da borda do marcador;</li>
</ul>

<br>

<p style="text-align: center; color:red; font-weight: bold; font-family: Times New Roman;">ATENÇÃO</p>

<br>

<p style="text-align: center; font-size: 18px; font-family: Times New Roman;">
Os parâmetros acima não funcionam corretamente para a versão 3.3.4 do matplotlib para plt.scatter. Foi verificado que funciona apenas para o plt.plot.
</p>


<br>
<p style="text-align: right; font-size: 12px; font-family: Times New Roman;"><a href="#topicos"><i>voltar ao índice</i></a></p>



```python
fig, (axs1, axs2) = plt.subplots(1,2)
fig.set_size_inches(16,8)
fig.suptitle("Comparando as edições")
axs1.plot(cachorro_peso, cachorro_altura, label="Cachorros", marker='o', markerfacecolor='m',  markeredgecolor='y')
axs1.plot(gatos_peso, gatos_altura, label="Gatos", marker='o', markerfacecolor='m',  markeredgecolor='y')
axs1.plot(pinguins_peso, pinguins_altura, label="Pinguins", marker='o', markerfacecolor='m',  markeredgecolor='y')
axs1.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Base")
axs1.legend()

axs2.plot(cachorro_peso, cachorro_altura, label="Cachorros", marker='o', markerfacecolor='m',  markeredgecolor='y')
axs2.plot(gatos_peso, gatos_altura, label="Gatos", marker='o', markerfacecolor='m',  markeredgecolor='y')
axs2.plot(pinguins_peso, pinguins_altura, label="Pinguins", marker='o', markerfacecolor='m',  markeredgecolor='y')
axs2.set(xlabel="Peso (kg)", ylabel="Altura (cm)", title="Alterando da cor da legenda baseado na cor do marcador")
axs2.legend(labelcolor='mfc')
plt.show()
```
